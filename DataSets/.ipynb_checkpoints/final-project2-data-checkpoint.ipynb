```python
# imports
import pandas as pd
import numpy as np
```

```python
# dataset for graduation by district and county
# removed dtype = 'object' because it was causing error down the road
district_data = pd.read_csv("District_Data_Download_4yr_2016.csv")
district_data
```

```python
# list of columns 
district_data.columns
```

```python
# extracted columns from dataframe
cnty_dist = district_data[['CNTYNAME', 'DIST_ALLR_GRAD']]
cnty_dist.head()
```

```python
# dropped any rows with 0.0 from our dataset.  Originally those rows had (dots), which was indicated by the dictionary section of
# the original excel file, that there were no students in that distric, therefore those rows were dropped  
# cnty_dist = cnty_dist[cnty_dist.DIST_ALLR_GRAD != 0.0 ]
```

```python
# By default pandas displays limited amount of rows, code below displays all rows in dataframe.
# print(district_data.to_string())

with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(cnty_dist)
```

```python
# grouped by county name and calculated average 
mean_grad = cnty_dist.groupby("CNTYNAME")["DIST_ALLR_GRAD"].mean()
# type(mean_grad)
mean_grad.head()
```

```python
# converted from series to dataframe
grad = mean_grad.to_frame()
grad.head()
```

```python
# reset dataframe index
index_reset = grad.reset_index()
index_reset.head()
print(index_reset.to_string())
```

```python
#changed name of columns for merge purposes 
index_reset.columns = ['County','Graduation Rate']
index_reset.head() 
```

```python
# remove the `County` and extra space (' County') from county column to match data from the income dataframe
index_reset['County'] = index_reset['County'].str.replace(' County','')
index_reset.head()
```

```python
# meant to strip extra space after deleting County from dataframe, but space was removed so not needed.
#index_reset['County'] = index_reset['County'].str.strip()
```

```python
income = pd.read_csv("county_income.csv")
# income.head()
print (income.to_string())
```

```python
# merge first two dataframes into one, income index_reset, income 
merged = pd.merge( index_reset, income, on='County', how ="outer" )
# merged.head()
print (merged.to_string())
```

```python
combined_data = pd.read_csv("combined_data.csv")
combined_data.head()
```

```python
combined_data.columns
```

```python
combined_data["Total SAT Score"] = combined_data["Total SAT Score"].replace("NAN", np.nan).astype(float)
combined_data.head()
```

```python
# extracted columns from dataframe
combined_ext = combined_data.loc[:, ['CntyName', 'Total SAT Score']]
combined_ext.head()
```

```python
# Rename of columns for merge purposes 
combined_ext.columns = ['County','SAT Score']
combined_ext.head() 
```

```python
#dropped county
combined_ext['County'] = combined_ext['County'].str.replace(' County','')
combined_ext.head()
```

```python
SAT = combined_ext.groupby("County").mean()
# SAT.head()
```

```python
# reset dataframe index
SAT = SAT.reset_index()
# SAT.head()
print (SAT.to_string())
```

```python
# added a third dataframe to the first two dataframes
all_data = pd.merge(merged, SAT, on='County', how="outer")
# all_data.head()
print (all_data.to_string())
```

```python
all_data.columns
```

```python
# renaming columns so there are no spaces
all_data.columns = ["County", "Graduation", "Income", "SAT"]
all_data.head()
```

```python
# saved cleaned data to a new csv
all_data.to_csv("../Datasets/merged.csv", index=False )
```
