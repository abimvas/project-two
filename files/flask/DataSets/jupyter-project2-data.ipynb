```python
# imports
import pandas as pd
```

```python
# dataset for graduation by district and county
# removed dtype = 'object' because it was causing error down the road
district_data = pd.read_csv("District_Data_Download_4yr_2016.csv")
district_data
```

```python
# list of columns 
district_data.columns
```

```python
# looped over dataframe to list all columns
district_data.columns
```

```python
# extracted columns from dataframe
cnty_dist = district_data[['CNTYNAME', 'DIST_ALLR_GRAD']]
cnty_dist.head()
```

```python
# drop any rows with 0.0 , from our data set those rows do not have any students and are therefore not relevant 
cnty_dist = cnty_dist[cnty_dist.DIST_ALLR_GRAD != 0.0 ]
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(cnty_dist)
```

```python
# grouped by county name and calculated average 
mean_grad = cnty_dist.groupby("CNTYNAME")["DIST_ALLR_GRAD"].mean()
# type(mean_grad)
mean_grad.head()
```

```python
# converted from series to dataframe
grad = mean_grad.to_frame()
grad.head()
```

```python
# reset dataframe index
index_reset = grad.reset_index()
index_reset.head()
```

```python
#changed name of columns for merge purposes 
index_reset.columns = ['County','Graduation Rate']
index_reset.head() 
```

```python
# remove the `County` and extra space (' County') from county column to match data from the income dataframe
index_reset['County'] = index_reset['County'].str.replace(' County','')
index_reset.head()
```

```python
# meant to strip extra space after deleting County from dataframe, but space was removed so not needed.
#index_reset['County'] = index_reset['County'].str.strip()
```

```python
income = pd.read_csv('county_income.csv')
income.head()
```

```python
# merge data, income index_reset, income  frames 
merged = pd.merge( index_reset, income, on='County' )
merged.head()
```

```python
# save cleaned data to csv
merged.to_csv("graduation_by_income.csv", index=False )
```
